```python
import re

# Attempt to read the CSV file with automatic separator detection
with open('chat.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    #print(lines)

    # Remove text within square brackets
    cleaned_lines = [re.sub(r'\[.*?\]', '', line) for line in lines]
    #print(cleaned_lines)
    # Remove the text "Kipa: " from each line
    cleaned_lines = [line.replace('Kipa: ', '') for line in cleaned_lines]
    cleaned_lines = [line.replace('Jani Rytk√∂nen: ', '') for line in cleaned_lines]
    cleaned_lines = [line.replace('Jussi: ', '') for line in cleaned_lines]
    print(cleaned_lines)








```

```python
from wordcloud import WordCloud

import matplotlib.pyplot as plt

# Join all the lines into a single string
text = ' '.join(cleaned_lines)

# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = stopwords.words("finnish")
```

```python
# Join all the lines into a single string
# Remove stop words from cleaned_lines
cleaned_lines = [' '.join([word for word in line.split() if word.lower() not in stop_words]) for line in cleaned_lines]

# Join all the lines into a single string
text = ' '.join(cleaned_lines)

# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
